# K-means

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 데이터 생성
data, labels = make_blobs(n_samples=300, centers=3, random_state=42)

# K-means 알고리즘 적용
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
predicted_labels = kmeans.labels_

# 결과 시각화
plt.scatter(data[:, 0], data[:, 1], c=predicted_labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red')
plt.title("K-means Clustering")
plt.show()
```

```python
# K-means 알고리즘 적용 (K=4)
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)
predicted_labels = kmeans.labels_

# 결과 시각화
plt.scatter(data[:, 0], data[:, 1], c=predicted_labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red')
plt.title("K-means Clustering with K=4")
plt.show()
```

# PCA

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# 데이터 로드
iris = load_iris()
data = iris.data
target = iris.target

# 데이터 표준화
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# PCA 적용
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

# 결과 시각화
plt.figure(figsize=(8, 6))
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=target, cmap='viridis', edgecolor='k', s=70)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D PCA of IRIS Dataset')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# 데이터 로드
digits = load_digits()
data = digits.data
target = digits.target

# PCA 적용 (3차원으로 축소)
pca = PCA(n_components=3)
principal_components = pca.fit_transform(data)

# 3D 시각화
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(principal_components[:, 0], principal_components[:, 1], principal_components[:, 2], c=target, cmap='viridis', edgecolor='k', s=70)
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D PCA of Digits Dataset')
ax.legend(*scatter.legend_elements(), title="Classes")
plt.show()
```

# Hierarchical Clustering

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import load_iris

# 데이터 로드
iris = load_iris()
data = iris.data

# 계층적 클러스터링
linked = linkage(data, method='single')

# 덴드로그램 시각화
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import make_moons

# 데이터 생성
data, labels = make_moons(n_samples=100, noise=0.05, random_state=42)

# 계층적 클러스터링
linked = linkage(data, method='complete')

# 덴드로그램 시각화
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Dendrogram for make_moons dataset")
plt.show()
```

# DASCAN

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# 데이터 생성
data, labels = make_moons(n_samples=300, noise=0.05, random_state=42)

# DBSCAN 적용
dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(data)

# 결과 시각화
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis', edgecolor='k', s=70)
plt.title("DBSCAN Clustering of make_moons dataset")
plt.show()
```

# 군집 알고리즘 비교

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.decomposition import PCA

# 데이터 생성
data, true_labels = make_moons(n_samples=300, noise=0.05, random_state=42)

# K-means
kmeans = KMeans(n_clusters=2)
kmeans_clusters = kmeans.fit_predict(data)

# PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(data)
kmeans_pca = KMeans(n_clusters=2)
pca_clusters = kmeans_pca.fit_predict(pca_data)

# 계층적 클러스터링
agg_clustering = AgglomerativeClustering(n_clusters=2, linkage='ward')
agg_clusters = agg_clustering.fit_predict(data)

# DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_clusters = dbscan.fit_predict(data)

# 결과 시각화
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Clustering Algorithm Comparisons")

axes[0, 0].scatter(data[:, 0], data[:, 1], c=kmeans_clusters, cmap='viridis', edgecolor='k', s=70)
axes[0, 0].set_title("K-means Clustering")

axes[0, 1].scatter(data[:, 0], data[:, 1], c=pca_clusters, cmap='viridis', edgecolor='k', s=70)
axes[0, 1].set_title("PCA followed by K-means")

axes[1, 0].scatter(data[:, 0], data[:, 1], c=agg_clusters, cmap='viridis', edgecolor='k', s=70)
axes[1, 0].set_title("Agglomerative Clustering")

axes[1, 1].scatter(data[:, 0], data[:, 1], c=dbscan_clusters, cmap='viridis', edgecolor='k', s=70)
axes[1, 1].set_title("DBSCAN")

plt.tight_layout()
plt.show()
```

DBSCAN은 비선형성을 띄는 데이터, 평균을 따르지 않는 데이터에 대한 군집을 잘 적용한다.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.decomposition import PCA

# 데이터 생성
data, true_labels = make_circles(n_samples=300, factor=0.5, noise=0.05, random_state=42)

# K-means
kmeans_clusters = KMeans(n_clusters=2).fit_predict(data)

# PCA
pca_data = PCA(n_components=2).fit_transform(data)
pca_clusters = KMeans(n_clusters=2).fit_predict(pca_data)

# 계층적 클러스터링
agg_clusters = AgglomerativeClustering(n_clusters=2, linkage='ward').fit_predict(data)

# DBSCAN
dbscan_clusters = DBSCAN(eps=0.3, min_samples=5).fit_predict(data)

# 결과 시각화
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("`make_circles` Clustering Algorithm Comparisons")

# 시각화 함수
def plot_clusters(ax, clusters, title):
    ax.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis', edgecolor='k', s=70)
    ax.set_title(title)

plot_clusters(axes[0, 0], kmeans_clusters, "K-means Clustering")
plot_clusters(axes[0, 1], pca_clusters, "PCA followed by K-means")
plot_clusters(axes[1, 0], agg_clusters, "Agglomerative Clustering")
plot_clusters(axes[1, 1], dbscan_clusters, "DBSCAN")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# 데이터 생성
data, labels = make_blobs(n_samples=500, centers=4, cluster_std=0.6, random_state=42)

# K-means 클러스터링 수행
kmeans = KMeans(n_clusters=4)
cluster_labels = kmeans.fit_predict(data)

# 실루엣 스코어 계산
silhouette_avg = silhouette_score(data, cluster_labels)
print(f"평균 실루엣 스코어: {silhouette_avg:.2f}")

# 각 데이터 포인트의 실루엣 스코어 계산
sample_silhouette_values = silhouette_samples(data, cluster_labels)

# 실루엣 스코어를 기반으로 시각화
fig, ax = plt.subplots(figsize=(8, 6))
y_lower = 10
for i in range(4):
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i
    ax.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values)
    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10

ax.axvline(x=silhouette_avg, color="red", linestyle="--")
ax.set_title("실루엣 스코어에 따른 클러스터 분포")
ax.set_xlabel("실루엣 계수 값")
ax.set_ylabel("클러스터 번호")
plt.show()
```
