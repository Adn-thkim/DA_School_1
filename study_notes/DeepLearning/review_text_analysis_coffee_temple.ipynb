SQL 환경 설정

상품명 별 리뷰의 수

```python
query = """
    SELECT 상품명, COUNT(리뷰상세내용) AS 리뷰수
    FROM product_reviews
    GROUP BY 1
    ORDER BY 2 DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
    SELECT 상품명,
           COUNT(리뷰상세내용) AS 리뷰수,
           AVG(구매자평점) AS 평균평점
    FROM product_reviews
    WHERE 상품명 LIKE '%쥬시%';
"""
pd.read_sql(query, conn)
```

커피 맛별 구매자 평점 집계

```python
query = """
    SELECT CASE WHEN 리뷰상세내용 LIKE '%쥬시%' THEN '쥬시'
                WHEN 리뷰상세내용 LIKE '%아메리카노%' THEN '아메리카노'
                WHEN 리뷰상세내용 LIKE '%에스프레소%' THEN '에스프레소'
                WHEN 리뷰상세내용 LIKE '%드립백%' THEN '드립백'
                WHEN 리뷰상세내용 LIKE '%캡슐%' THEN '캡슐'
                ELSE '기타'
           END AS 커피종류,
           AVG(구매자평점) AS 평균평점
    FROM product_reviews
    GROUP BY 1;
"""
pd.read_sql(query, conn)
```

평점 4 이상의 상품 중에서 '에스프레소'라는 맛이 들어간 리뷰의 상품명과 평점을 찾기

```python
query = """
    SELECT CASE WHEN 리뷰상세내용 LIKE '%쥬시%' THEN '쥬시'
                WHEN 리뷰상세내용 LIKE '%아메리카노%' THEN '아메리카노'
                WHEN 리뷰상세내용 LIKE '%에스프레소%' THEN '에스프레소'
                WHEN 리뷰상세내용 LIKE '%드립백%' THEN '드립백'
                WHEN 리뷰상세내용 LIKE '%캡슐%' THEN '캡슐'
                ELSE '기타'
           END AS 커피종류,
           AVG(구매자평점) AS 평균평점
    FROM product_reviews
    WHERE 리뷰상세내용 LIKE '%에스프레소%'
    GROUP BY 1
    HAVING 평균평점>=4;
"""
pd.read_sql(query, conn)
```

'아메리카노'라는 맛을 포함한 리뷰 중 가장 오래된 리뷰 찾기

```python
query = """
    SELECT 리뷰상세내용,
           SUBSTR(리뷰등록일, 1, 10) AS 일자
    FROM product_reviews
    WHERE 리뷰상세내용 LIKE '%아메리카노%'
    ORDER BY 2
    LIMIT 1;
"""
pd.read_sql(query, conn)
```
