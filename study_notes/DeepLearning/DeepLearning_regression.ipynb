```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([80, 90, 88, 92])
```

```python
a = 0  # 기울기
b = 0  # 절편

lr = 0.03  # Learning Rate 학습률

epochs = 1000
n = len(x)
```

```python
# Gradient Decsent
for i in range(epochs):
    y_pred = a * x + b
    error = y - y_pred

    a_diff = (2/n) * sum(-x * (error))  # 오차 함수를 a로 편미분한 값
    b_diff = (2/n) * sum(-(error))  # 오차 함수를 b로 편미분한 값

    a = a - lr * a_diff  # a 업데이트
    b = b - lr * b_diff  # b 업데이트

    if i % 50 == 0:
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b))

y_pred = a * x + b

plt.scatter(x, y)
plt.plot(x, y_pred,'r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([[2, 3], [4, 4], [6, 2], [8, 5]])
y = np.array([80, 90, 88, 92])

model = Sequential()

# Layer 추가
model.add(Dense(1, input_dim=2, activation='linear')) # 1개의 uniut, input은 2차원 배열
# input_shape을 통해 초기 가중치가 지정되고, 모델이 입력 데이터의 차원을 알아서 잘 적용하지 못함
# 따라서, 초기 가중치가 제대로 설정되지 않으면 학습이 제대로 이뤄질 수 없다.
# activation='linear'와 activation=None은 둘 다 선형 레이어로서 동작하며, 비선형성을 추가하지 않는다.
# 출력 : 입력에 가중치를 곱하고 편향을 더한 결과
model.compile(optimizer='sgd', loss='mse')

model.fit(x, y, epochs=2000)
```

```python
# plt.scatter(x, y)
# plt.plot(x, model.predict(x), 'r')
# plot.show()
```

```python
hour = 7
private_class = 4

prediction = model.predict([[hour, private_class]])

print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점입니다" % (hour, private_class, prediction))
```

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([[2, 3], [4, 4], [6, 2], [8, 5]])
y = np.array([80, 90, 88, 92])

model = Sequential()

# Layer 추가
model.add(Dense(1, input_dim=2))
model.compile(optimizer='sgd', loss='mse')

model.fit(x, y, epochs=2000)
```

```python
hour = 7
private_class = 4

prediction = model.predict([[hour, private_class]])

print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점입니다" % (hour, private_class, prediction))
```

기본적인 모델 설계

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

np.random.seed(0)

X_example = np.random.rand(100, 3)
Y_example = np.random.randint(2, size=100)

X_example[:5], Y_example[:5]
```

```python
model_example = Sequential()
model_example.add(Dense(16, input_dim=3, activation='relu'))  # units 16인 Hidden Layer
model_example.add(Dense(1, activation='sigmoid'))  # Output Layer

model_example.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model_example.fit(X_example, Y_example, epochs=100, verbose=0)
# verbose: 출력되는 정보의 양 조절
# verbose=0: 출력 없음
# verbose=1: 진행 바(default) / 각 epoch마다 진행 상황, Loss, Metric 표시
# verbose=2: 간결한 출력 / 각 epoch의 진행 상황을 출력하되, 진행 바가 표시되지 않음. Loss, Metric만 출력
```

```python
loss, accuracy = model_example.evaluate(X_example, Y_example)
print(accuracy)
```
