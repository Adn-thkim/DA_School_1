```python
import pandas as pd

reviews = pd.read_excel('review_20230725_091330.xlsx')
reviews.head()
```

```python
content = ' '.join([i for i in reviews['리뷰상세내용']])
content
```

```python
from konlpy.tag import Okt

nlp = Okt()
content_N = nlp.nouns(content)
content_N
```

```python
from collections import Counter

count = Counter(content_N)
count
```

```python
word_count = dict()
for tag, counts in count.most_common(80):
  if (len(str(tag))>1):
    word_count[tag] = counts
    print("%s : %d" %(tag,counts))
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
plt.figure(figsize = (12,5))
plt.xlabel('키워드')
plt.ylabel('빈도수')
plt.grid(True)
sorted_Keys = sorted(word_count, key = word_count.get, reverse = True)
sorted_Values = sorted(word_count.values(), reverse = True)
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='vertical')
plt.show()
```
