# 선 그래프

그래프를 출력하는 기본적인 형태
```python
import matplotlib.pyplot as plt 

plt.plot(data)
plt.show()
```

**plot**: plot(x1, y1, fmt, x2, y2, fmt, ...) 형태의 입력을 받아 그래프를 생성한다
- 리스트([1, 2, 3, 4]) 한 개를 입력하면, y값들이라고 가정하고 x값 [0, 1, 2, 3] (idx)을 자동으로 지정한다
**show**: 그래프를 화면에 출력한다

````python
import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4])
plt.show()
````

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
````

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], fillstyle = 'full')
plt.show()
````

## Line, marker

plot 함수의 parameter 설정을 통해 선과 마커의 종류, 색상 등을 설정할 수 있다
- fmt: 포맷 문자열을 입력받아 그래프 선의 형태 지정
    - fmt = '[marker] [line] [color]' ex. -ob : 원형의 marker가 있는, 파란색 실선
- marker: 사용할 marker의 형태 지정
- color: 사용할 color 지정
- linestyle: linestyle(실선, 점선 등) 지정
- capstyle: 선의 모서리 모양(butt: 각진 모서리, round: 둥근 모서리) 지정
    - linestyle이 solid라면, parameter를 solid_capstyle로 사용
- joinstyle: 두 선이 만나는 부분의 모양(miter: , round, bevel) 지정
- linewidth: 선의 굵기 지정

튜플을 사용해서 선의 종류를 커스터마이즈할 수 있다

예를 들어, (0, (1, 1))은 ‘dotted’와 같고, (0, (5, 5))는 ‘dashed’와 같다

또한 (0, (3, 5, 1, 5))는 ‘dashdotted’와 같다

````python
plt.plot([1, 2, 3], [4, 4, 4], linestyle=(0, (1, 1)), color='C0', label='(0, (1, 1))')
plt.plot([1, 2, 3], [3, 3, 3], linestyle=(0, (1, 5)), color='C0', label='(0, (1, 5))')
plt.plot([1, 2, 3], [2, 2, 2], linestyle=(0, (5, 1)), color='C0', label='(0, (5, 1))')
plt.plot([1, 2, 3], [1, 1, 1], linestyle=(0, (3, 5, 1, 5)), color='C0', label='(0, (3, 5, 1, 5))')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
````

````python
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', linewidth=10,
      solid_capstyle='butt', color='C0', label='solid+butt')
plt.plot([1, 2, 3], [3, 3, 3], linestyle='solid', linewidth=10,
      solid_capstyle='round', color='C0', label='solid+round')

plt.plot([1, 2, 3], [2, 2, 2], linestyle='dashed', linewidth=10,
      dash_capstyle='butt', color='C1', label='dashed+butt')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashed', linewidth=10,
      dash_capstyle='round', color='C1', label='dashed+round')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
````

# 축 설정

## xlabel, ylabel

````python
# 레이블이 있는 데이터 사용하기

import matplotlib.pyplot as plt

font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }

data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}

plt.plot('data_x', 'data_y', data = data_dict)
plt.xlabel(list(data_dict.keys())[0], loc = 'right', labelpad = 15, fontdict = font1)
plt.ylabel(list(data_dict.keys())[1], loc = 'top', labelpad = 20, fontdict = font2)
plt.show()
````

## axis, xlim, ylim

**axis**: [xmin, xmax, ymin, ymax]를 입력으로 받아 x축, y축의 범위를 지정한다
- axis는 x, y축의 범위를 동시에 지정
- xlim으로 x축의 범위만 설정, ylim으로 y축의 범위만 설정할 수 있다
- axis('scaled'), xlim(), ylim() 함수로 축의 범위를 반환할 수 있다

````python
# plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'bo-')    # 파란색 + 마커 + 실선
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'bo--')     # 파란색 + 마커 + 점선
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
````

````python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'bd--')     # 파란색 + 마커 + 점선
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
````

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # 포맷 문자열 ‘ro’는 빨간색 (‘red’)의 원형 (‘o’) 마커를 의미
                                            # 예를 들어 ‘b-‘는 파란색 (‘blue’)의 실선 (‘-‘)을 의미
plt.axis([0, 6, 0, 20])                     # [xmin, xmax, ymin, ymax]
plt.show()
````

````python
# 축범위 얻기 
# xlim(), ylim() 함수는 그래프 영역에 표시되는 X축, Y축의 범위를 각각 반환합니다.
# axis() 함수는 그래프 영역에 표시되는 X, Y축의 범위를 반환합니다.

import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

x_range, y_range = plt.xlim(), plt.ylim()
print(x_range, y_range)

axis_range = plt.axis('scaled')
print(axis_range)

plt.show()
````

## 그래프 영역 설정

fill_between: 특정 영역을 강조하기 위해 사용
- fillstyle: {'full', 'left', 'right', 'bottom', 'top', 'none'}

fill_between() 함수에 x[1:3], y[1:3]를 순서대로 입력하면,

네 점 (x[1], y[1]), (x[2], y[2]), (x[1], 0), (x[2], 0)을 잇는 영역이 채워진다.

````python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)

plt.show()
````

# 막대 그래프

bar 함수의 기본적인 형태 bar(x, height)

- color
- width
- edgecolor
- linewidth
- tick_label
- align

아래 예제는 연도별로 변화하는 값을 갖는 데이터를 막대 그래프로 나타냅니다.

[0,1,2] 주어진 범위와 간격에 따라 균일한 값을 갖는 리스트를 생성합니다.

years는 X축에 표시될 연도이고, values는 막대 그래프의 y 값 입니다.

먼저 plt.bar() 함수에 x 값 [0, 1, 2]와 y 값 [100, 400, 900]를 입력해주고,

xticks()에 x와 years를 입력해주면, X축의 눈금 레이블에 ‘2018’, ‘2019’, ‘2020’이 순서대로 표시됩니다.

결과는 아래와 같습니다.

````python
x = [0, 1, 2]
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)

plt.show()
````

````python
x = [0,1,2]
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values, color='#e35f62')
plt.xticks(x, years)

plt.show()
````

````python
# 값별로 색상 다르게 하기

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.bar(x, values, color=colors)
plt.xticks(x, years)

plt.show()
````

````python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.bar(x, values, color = colors, width = 1.0)
plt.xticks(x, years)

plt.show()
````

````python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.bar(x, values, align='edge', edgecolor='lightgray',
        linewidth=5, tick_label=years)

plt.show()
````

````python
fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
````

````python
import csv
import os

data_path = '/Users/thkim/Documents/likelion/source/sample_data'

f = open(os.path.join(data_path, 'CSV_example2_data.csv'), 'r') #csv를 리드 모드로 불러옴
data = csv.reader(f)
next(data)

result_april = []
result_may = []

for row in data: 
    month = row[0].split('-')[1]
    if month == '04':
        result_april.append(float(row[-1]))
    elif month == '05':
        result_may.append(float(row[-1]))

f.close()
````

문제 4. 제주도의 남성 0-100세, 여성 0-100세까지 연령에 따른 막대그래프를 출력하세요

````python
f = open(os.path.join(data_path, 'dataset_example3_new.csv'), 'r') #csv를 리드 모드로 불러옴
data = csv.reader(f)
header = next(data) # 헤더 정보 추출

female = []
male = []

m_idx_s = header.index('남_0-9세')
m_idx_f = header.index('남_100세 이상')

f_idx_s = header.index('여_0-9세')
f_idx_f = header.index('여_100세 이상')

for row in data:
    if row[0] == '제주특별자치도':
        for m in row[m_idx_s:m_idx_f]:
            male.append(int(m.replace(',', '')))
        for fe in row[f_idx_s:f_idx_f]:
            female.append(int(fe.replace(',', '')))
    
f.close()
````

````python
plt.bar(range(len(male)), male, label = 'male')
plt.bar(range(len(female)), female, label = 'female')
plt.legend()
plt.show()
````

````python
plt.barh(range(len(male)), male, label = 'male')
plt.barh(range(len(female)), list(map(lambda x: 0 - x, female)), label = 'female')
plt.legend()
plt.show()
````

문제 5. 연도 구분별 서울과 수도권의 인구 증감률 데이터 'example_4_all.csv'를 이용해서 '구분', '서울', '수도권'을 나누어 리스트를 생성하세요

````python
year = []
s = []
g = []

with open(os.path.join(data_path, 'example_4_all.csv')) as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        year.append(row[0])
        s.append(float(row[1]))
        g.append(float(row[2]))
````
