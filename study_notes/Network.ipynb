- **자체 루프 및 다중 엣지 불허**: **`nx.Graph()`**로 생성된 그래프는 노드 자신에게 향하는 엣지(자체 루프)나, 두 노드 사이에 여러 개의 엣지(다중 엣지)를 허용하지 않습니다.

### 4. **커뮤니티 탐지 및 결과 해석**

커뮤니티는 네트워크 내에서 밀접하게 연결된 노드 그룹을 의미합니다.

- NetworkX에서는 Louvain 방법과 같은 다양한 커뮤니티 탐지 방법을 사용할 수 있습니다.
- Zachary's karate club 데이터셋에서는 여러개의 커뮤니티를 확인할 수 있습니다.

**Louvain 방법**은 대규모 네트워크의 커뮤니티 구조를 탐지하기 위한 알고리즘 중 하나입니다. 이 방법은 다음의 주요 원칙에 기반합니다:

1. **모듈성 최적화**: Louvain 방법의 핵심 목표는 네트워크의 모듈성을 최적화하는 것입니다. 여기서 모듈성이란 네트워크 내의 노드 그룹(커뮤니티)이 얼마나 밀접하게 연결되어 있는지를 측정하는 지표입니다. 높은 모듈성은 해당 커뮤니티 내부의 엣지가 많고 외부와의 엣지는 적다는 것을 의미합니다.
2. **계층적 접근법**: 이 알고리즘은 여러 단계를 거쳐서 수행됩니다. 첫 번째 단계에서는 각 노드를 개별 커뮤니티로 간주합니다. 그 후, 각 노드를 인접한 커뮤니티 중에서 모듈성을 가장 크게 증가시키는 커뮤니티로 이동시킵니다. 이 단계가 더 이상 모듈성을 향상시키지 못할 때까지 반복됩니다.
3. **커뮤니티 병합**: 한 번의 최적화가 완료되면, 각 커뮤니티는 하나의 '슈퍼 노드'로 간주되어 새로운 네트워크를 형성합니다. 이 과정은 새로운 네트워크에 대해 모듈성이 더 이상 향상되지 않을 때까지 반복됩니다.

결론적으로, Louvain 방법은 네트워크 내의 커뮤니티 구조를 탐지하기 위해 모듈성 최적화를 중심으로 하는 계층적 접근법을 사용하는 알고리즘입니다. 이 방법은 대규모 네트워크에서도 효과적이며, 다양한 네트워크 분석 분야에서 널리 사용됩니다.

- **`community_louvain.best_partition(G)`** 함수는 그래프 **`G`**의 노드들을 최적의 커뮤니티로 분류합니다.
- **`partition`**은 딕셔너리 형태로 반환되며, 각 노드를 키로, 해당 노드가 속한 커뮤니티의 ID를 값으로 갖습니다.

## **Girvan-Newman 알고리즘:**

**Girvan-Newman 알고리즘**은 네트워크에서 커뮤니티를 탐지하기 위한 방법 중 하나입니다. 이 알고리즘의 주요 원리는 "매개 중심성"을 기반으로 엣지를 제거하면서 커뮤니티를 분리하는 것입니다. 여기에는 몇 가지 핵심 단계가 있습니다:

1. 모든 엣지의 매개 중심성을 계산합니다.
2. 매개 중심성이 가장 높은 엣지를 네트워크에서 제거합니다.
3. 이 과정을 반복하면서 커뮤니티가 분리되는 시점을 관찰합니다.

이 알고리즘의 핵심 아이디어는 네트워크 내에서 많은 최단 경로들이 통과하는 엣지가 중요한 "브릿지" 역할을 한다는 것이며, 이러한 엣지를 제거함으로써 커뮤니티를 분리할 수 있다는 것입니다.

```python
from networkx.algorithms.community import girvan_newman

# Girvan-Newman 알고리즘을 사용한 커뮤니티 탐지
communities = girvan_newman(G_florentine)
top_level_communities = next(communities)
community_groups = sorted(map(sorted, top_level_communities))

community_groups
```

# 실습 - 공급망(운송비용) 최적화 문제

**시나리오:**

**아이템 A**는 여러 공급자로부터 구입되어 중앙 창고에 저장되며, 이 중앙 창고에서 여러 고객에게 배송됩니다. 각 공급자와 중앙 창고, 그리고 중앙 창고와 고객 사이의 운송 비용이 다릅니다. 우리의 목표는 총 운송 비용을 최소화하는 방법을 찾는 것입니다.

**데이터:**

1. 공급자: S1, S2, S3
2. 고객: C1, C2, C3
3. 중앙 창고: W

각 공급자에서 W로, 그리고 W에서 각 고객으로의 운송 비용은 다음과 같습니다:

- S1 -> W: \$5
- S2 -> W: \$7
- S3 -> W: \$6
- W -> C1: \$2
- W -> C2: \$4
- W -> C3: \$3

**문제 해결:**

이 문제의 경우, 각 공급자에서 아이템을 구입하는 비용이 동일하므로, 최소 비용의 경로만 선택하면 됩니다. 공급업체(supplier), 창고(warehouse), 그리고 고객(customer) 간의 무게(비용)를 가진 방향성 그래프를 생성합니다. 주요 목적은 최소 비용 경로를 식별하는 것입니다.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
suppliers = ['S1', 'S2', 'S3']
customers = ['C1', 'C2', 'C3']
warehouse = 'W'

G.add_nodes_from(suppliers, role='supplier')
G.add_nodes_from(customers, role='customer')
G.add_node(warehouse, role='warehouse')

# Add edges with weights (costs)
G.add_edge('S1', 'W', weight=5)
G.add_edge('S2', 'W', weight=7)
G.add_edge('S3', 'W', weight=6)
G.add_edge('W', 'C1', weight=2)
G.add_edge('W', 'C2', weight=4)
G.add_edge('W', 'C3', weight=3)

# Identify the minimum cost paths
min_cost_supplier = min([(data['weight'], supplier) for supplier, _, data in G.edges(data=True) if G.nodes[supplier]['role'] == 'supplier'])[1]
min_cost_customer = min([(data['weight'], customer) for _, customer, data in G.edges(data=True) if G.nodes[customer]['role'] == 'customer'])[1]

min_cost_supplier, min_cost_customer
```

```python
# Plot the network
plt.figure(figsize=(12, 8))
pos = {
    'S1': (1, 2), 'S2': (1, 1), 'S3': (1, 0),
    'W': (2, 1),
    'C1': (3, 2), 'C2': (3, 1), 'C3': (3, 0)
}

# pos 딕셔너리는 각 노드의 (x, y) 위치를 나타냅니다. 이를 통해 그래프를 그릴 때 노드의 위치를 정의합니다.
# 최소 비용의 공급업체에서 창고로, 그리고 창고에서 최소 비용의 고객으로 향하는 엣지는
# 빨간색과 두꺼운 너비로 표시되며, 나머지 엣지는 검은색과 얇은 너비로 표시됩니다.

edge_colors = ['red' if (u == min_cost_supplier and v == 'W') or (u == 'W' and v == min_cost_customer) else 'black' for u, v in G.edges()]
edge_widths = [2 if (u == min_cost_supplier and v == 'W') or (u == 'W' and v == min_cost_customer) else 1 for u, v in G.edges()]
# 각 엣지의 무게(비용) 값을 레이블로 사용하기 위해 nx.get_edge_attributes 함수를 사용하여 무게 값을 가져옵니다.
labels = nx.get_edge_attributes(G, 'weight')

nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', edge_color=edge_colors, width=edge_widths, font_size=15)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=15)
plt.title("Supply Chain Network")
plt.show()
```

위의 그래프에서 볼 수 있듯이:

- **빨간색**의 강조된 경로는 최소 비용의 경로를 나타냅니다.
- 각 화살표 옆의 숫자는 해당 경로의 운송 비용을 나타냅니다.

이 결과를 바탕으로, 아이템을 **S1** 공급자로부터 구매하고, 중앙 창고에서 **C1** 고객에게 배송하면 총 운송 비용을 최소화할 수 있습니다.

이러한 분석은 기업이 물류 및 배송 비용을 줄이기 위한 전략을 세울 때 매우 유용합니다. 물론, 실제 상황에서는 다양한 제약사항, 다양한 비용, 수요 예측 등 많은 요소들이 고려되어야 하지만, 이 예제는 간단한 최적화 문제를 어떻게 접근할 수 있는지를 보여주기 위한 것입니다.
